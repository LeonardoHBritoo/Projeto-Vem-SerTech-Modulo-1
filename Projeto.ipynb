```python
#Cria o menu de exibição
def menu():
    menu = '''
  ---------- Menu ----------
  [1] - Adicionar Restaurante
  [2] - Realizar Pedido
  [0] - Finalizar
  --------------------
  => '''
    return input(menu) #Retorna o valor recebido por input

#Retornar, em uma função, significa tornar aquele valor disponível em outras partes do código
```

```python
#Função para adicionar pratos no cardápio
#Deve ser inicializada antes da função adiciona_restaurante, mesmo ela sendo executada depois
#Pois seu conteúdo é utilizado durante a execução da função adiciona_restaurante
def adiciona_cardapio():
  cardapio = [] #cria uma lista de escopo local para o cardápio
  finaliza = True #Cria uma variável com valor verdadeiro para iniciar o loop (Kuma nunca aprovaria isso)
  while finaliza: #enquanto finaliza for verdadeirao
    prato = input('Nome do prato: ') #Recebe o nome do prato
    valor_prato_str = input('Valor do prato: ') #Recebe o preço do prato
    valor_prato_str = valor_prato_str.replace(',', '.') #Trata os erros se o usuário colocar vírgula (Copiei do Cesar, na cara dura mesmo)
    valor_prato = float(valor_prato_str) #Depois de garantir que o valor pode ser float, tranforma em float
    tempo_preparo = input('Tempo de preparo: ') #Recebe o tempo de preparo do prato, é necessário tratar erro aqui?
    print('[0] - Finalizar \n[1] - Continuar') #Exibe um menuzinho de escolha (Gosto de deixar claro o que tá acontecendo na interface)
    finalizar = input() #Recebe o valor de continuar ou finalizar
    finaliza = finalizar != '0' #Se o usuário digitar 0, vai sair do loop(Qualquer outra coisa, continua. Seria um problema?)
    pratos = [] #Cria uma lista dentro do loop para juntar as informações do prato
    pratos.append(prato) #Adiciona os valores à lista. Tem que ser um de cada vez
    pratos.append(valor_prato)
    pratos.append(tempo_preparo)
    cardapio.append(pratos) #No final, adiciona a lista pratos à lista cardapio. Ao reiniciar o loop, pratos estará vazia
  return cardapio #Retorna a lista cardápio, para ser usada na função adiciona restaurante
```

```python
#Função para adicionar restaurantes
def adiciona_restaurante():
  restaurante = input('Nome do restaurante: ') #Recebe o nome do restaurante
  cardapio = adiciona_cardapio() #Chama a função de adicionar cardápio
  restaurante_atual = [] #Cria uma lista local que armazena os valores para cada restaurante
  restaurante_atual.append(restaurante) #Adiciona o nome do restaurante à lista
  restaurante_atual.append(cardapio) #Adiciona a lista da função adiciona_cardapio à lista restaurante_atual
  restaurantes.append(restaurante_atual) #Por fim, adiciona a lista restaurante local a lista de restaurantes

#Restaurante_atual é criada a cada inserção de restaurantes, o que permite separá-los na lista restaurantes
#Como a lista restaurante_atual tem escopo local, ela encerra ao final da função sem armazenar dados
```

```python
#Função que exibe o menu (cardápio)
def exibe_menu(nome_restaurante): #Recebe o nome do restaurante como parâmetro
  for item in restaurantes: #Para cada item da (sub)lista
    if nome_restaurante == item[0]: #Se o nome buscado for igual ao nome do restaurante (lista posição 0)
      #Cabeçalho
      print(f'            Cardápio do Restaurante {item[0]}\n')
      print('|     Prato       |      Preço       | Tempor de Preparo  |')
      print('-'*58)
      prato = item[1] #Não sei um nome bom pra colocar aqui, se alguém descobrir, agradeço
      for p in prato: #Para cada item em prato
        print(f'|{p[0]:^16} | {p[1]:^16} | {p[2]:^18} |')
        print('-'*58)
```

```python
#Função para exibir pedido
#Foi declarada antes, pois a função realizar_pedido a utiliza
def exibir_pedido(pedido):
  #Ordenando a lista
  ordem = sorted(pedido)
  #Lista para captação de repetições
  tem = []
  print('Nº |     Prato       |      Preço       | Tempo de Preparo  | Quantidade | ')
  #Loop para contagem e impressão de valores
  for i in ordem:
    qnt = ordem.count(i) #conta o número de repetições e agrega à variável
    if i not in tem: #Se o elemento estiver passando pela primeira vez
    #Multiplico i[1] por qnt, ou seja, o preço pela quantidade, assim já sai o valor final de cada prato
      print(f'{pedido.index(i):02d} | {i[0]:^15} | {(i[1] * qnt):^16} | {i[2]:^17} | {qnt:^11}|') #Imprime no esquema do Leonardo. Ficou maneiro
      tem.append(i) #Adiciona o elemento à lista de repetições
    else: #Se o elemento já estiver na lista de repetições, pule-o e siga com o loop
      continue
```

```python
#Função para realização de pedido
#Foi criada aqui, pois é utilizada na função selecionar_restaurante
def realizar_pedido():
  #Optei por manter os pedidos restritos a execução. Não sei se seria interessante mantê-los posteriormente
  pedidos = [] #Cria uma lista vazia
  sem_ideia = 1 #Variável para manter o loop. O nome tá horrível, mas tá funcionando
  while sem_ideia:
    pedido = input('Digite o prato que desejado: ').title() #Pede o valor e transforma as primeiras letras em maiúscula
    for item in restaurantes:
      prato = item[1] #Variável prato possui valor de item na posição 1
      for p in prato: #Coloquei p pq não consegui pensar em nada
        if pedido in p: #Se o input do usuário for um prato existente
          pedidos.append(p) #Adiciona as informações do prato em pedidos
    print('[1] - Continuar com pedido \n[0] - Finalizar pedido')
    continua_pedido = input() #Testa o valor da variável que mantém o loop
    if continua_pedido == "0":
      sem_ideia = 0 #0 é igual a falso, isso quebra o loop
  exibir_pedido(pedidos) #Chama a função exibir_pedidos, passando a lista de pedidos como parâmetro

#Alguém altera o input de adiciona_cardapio com title(), capitalize(), upper() etc
#Assim os valores já vão estar padronizados na lista
#Só não esquece de atualizar aqui também (caso não use title()). Na variável pedido, que recebe input

#Atualmente, para pedir a mesma coisa duas vezes é necessário digitar o nome duas vezes (no exemplo fica mais visivel)
#Seria legal uma forma que o usuário digite quantos ele quer
#Por exemplo, Lasagna, 3. (Ou seja 3 lasagnas). Ao invés de digitar: lasagna, lasagna, lasagna
```
